```python
# data type : series
income = np.log10(app_train['AMT_INCOME_TOTAL'])
credit = np.log10(app_train['AMT_CREDIT'])
annuity = np.log10(app_train['AMT_ANNUITY'])
price = np.log10(app_train['AMT_GOODS_PRICE'])
tmp_dict = {
    'AMT_INCOME_TOTAL':income,
    'AMT_CREDIT':credit,
    'AMT_ANNUITY':annuity,
    'AMT_GOODS_PRICE':price
}

df = pd.DataFrame(tmp_dict)
df.plot(kind='hist', bins=50, alpha=0.5)
df.plot(kind='hist', bins=100, alpha=0.5)
df.plot(kind='hist', bins=120, alpha=0.5)
df.plot(kind='hist', bins=130, alpha=0.5)

# >>> df4.plot.hist(stacked=True, bins=20, alpha=0.5) 
# bins=20表示数值分辨率，具体来说是将随机数设定一个范围，
# 例如5.6，5.7，6.5，如果数值分辨率越低，则会将三个数分到5-7之间，
# 数值分辨率越高，则会将5.6，5.7分到5-6之间，而6.5分到6-7之间；
# 值越小表示分辨率越低，值越大表示分辨率越高；
ref = '''
    作者：genome_denovo 
    来源：CSDN 
    原文：https://blog.csdn.net/genome_denovo/article/details/78322628 
    版权声明：本文为博主原创文章，转载请附上博文链接！
'''
```

```python
weekday = pd.Series(app_train['WEEKDAY_APPR_PROCESS_START'])
print(weekday)

weekday.plot(kind='hist')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(app_train['WEEKDAY_APPR_PROCESS_START'])
weekday = le.transform(app_train['WEEKDAY_APPR_PROCESS_START'])
print('type : ', type(weekday), '\ndata : ', weekday, '\n')
weekday = pd.Series(weekday)
weekday.plot(kind='hist', bins = 30)
plt.xticks(np.arange(0,7), ['Sun','Mon','Tue','Wed','Thu','Fri','Sat'] )
# plt.show()
```
